```python
# Merge sales with customers on 'CustomerKey'
merged_df = pd.merge(sales, customers, on='CustomerKey', how='left')

# Merge with products on 'ProductKey'
merged_df = pd.merge(merged_df, products, on='ProductKey', how='left')

# Merge with stores on 'StoreKey'
merged_df = pd.merge(merged_df, stores, on='StoreKey', how='left')

# Merge with exchangerates on matching 'Order Date' with 'Date' (for currency conversion rates)
# Merge with exchangerates on matching 'Order Date' with 'Date' and 'Currency Code' with 'Currency'
merged_df = pd.merge(
    merged_df,
    exchangerates,
    left_on=['Order Date','Currency Code'],  # Use a list for left_on
    right_on=['Date','Currency'],             # Use a list for right_on
    how='left'
)
# Drop the 'Date' column from the exchange rates table as it might duplicate the 'Order Date'
merged_df = merged_df.drop(columns=['Date'])
# Rename the conflicting columns after merging
merged_df.rename(columns={
    'Country_x': 'CustomerCountry',  # Rename store country
    'State_x': 'CustomerState',      # Rename store state
    'Country_y': 'StoreCountry',  # Rename customer country
    'State_y': 'StoreState'   # Rename customer state
}, inplace=True)

# Now print the DataFrame
#print(merged_df.head())


# Final merged data
allinone=merged_df
allinone
```

```python
def colrename(x):
    colum=list(x.columns)
    colum1=[]
    for c in colum:
        colum1.append(c.replace(' ','_').lower())
    
    return colum1
```

```python
import pandas as pd
import pymysql

# Establish the connection
myconnection = pymysql.connect(host='localhost', user='root', passwd='Vinodha@123', database="datasparkproject")
cursor = myconnection.cursor()

# Read the CSV file
allinone = pd.read_csv("C:/Users/ncssa/Downloads/newallinone.csv")
customers = pd.read_csv("C:/Users/ncssa/Downloads/newcustomers.csv")
sales = pd.read_csv("C:/Users/ncssa/Downloads/newsales.csv")
stores = pd.read_csv("C:/Users/ncssa/Downloads/newstores.csv")
products = pd.read_csv("C:/Users/ncssa/Downloads/newproducts.csv")
exchangerates = pd.read_csv("C:/Users/ncssa/Downloads/newexchangerates.csv")

def table_creation(dataframe, tablename):
    # Mapping Pandas dtypes to SQL data types
    dtype_mapping = {
        'float64': 'FLOAT',
        'int64': 'INT',
        'object': 'TEXT',
        'category': 'TEXT',
        'datetime64[ns]': 'DATE'
    }

    # Generate column definitions for SQL
    columns = ", ".join(f"{col} {dtype_mapping[str(dtype)]}" for col, dtype in zip(dataframe.columns, dataframe.dtypes))

    # Create the table if it doesn't exist
    cursor.execute(f"CREATE TABLE IF NOT EXISTS {tablename} ({columns})")

    # Insert data into the table
    for i in range(len(dataframe)):
        row = tuple(dataframe.iloc[i])
        placeholders = ', '.join(['%s'] * len(row))  # Generate placeholders for SQL values
        sql = f"INSERT INTO {tablename} VALUES ({placeholders})"
        cursor.execute(sql, row)

# Call the function to create table and insert data
table_creation(allinone, 'allinone')
print("allinone table successfully created")
table_creation(customers, 'customers')
print("customers table successfully created")
table_creation(sales, 'sales')
print("sales table successfully created")
table_creation(stores, 'stores')
print("stores table successfully created")
table_creation(products, 'products')
print("products table successfully created")
table_creation(exchangerates, 'exchangerates')
print("exchangerates table successfully created")

# Commit the transaction
myconnection.commit()

# Close the cursor and connection
cursor.close()
myconnection.close()

print("Done")
```
